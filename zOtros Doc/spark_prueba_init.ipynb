```scala
import $ivy.`org.apache.spark::spark-sql:4.1.1`
import org.apache.spark.sql.SparkSession
```

```scala
println(s"Sesión de Spark iniciada. Versión: ${spark.version}")
```

```scala
val df = spark.createDataFrame(Seq(
(1, "Scala"), 
(2, "Spark"), 
(3, "Jupyter")
)).toDF("id", "herramienta")
df.show()
```
