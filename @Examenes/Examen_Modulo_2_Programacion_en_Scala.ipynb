# 🧪 Pregunta 8: Análisis de temperaturas semanales

Una estación meteorológica registra la **temperatura media semanal** (en grados Celsius) durante un trimestre (12 semanas):

```scala
val temperaturas = List(
  12.5, 13.1, 14.8, 15.2,
  16.0, 18.3, 19.1, 20.0,
  18.7, 17.2, 15.9, 14.4
)

val semanas = List(
  "Sem 1", "Sem 2", "Sem 3", "Sem 4",
  "Sem 5", "Sem 6", "Sem 7", "Sem 8",
  "Sem 9", "Sem 10", "Sem 11", "Sem 12"
)
```

Tareas:

- Usa `zip` para emparejar cada semana con su temperatura.
- Imprime las **3 semanas con mayor temperatura** (usa `sortBy` y `take`).
- Calcula e imprime:
  - la **temperatura media del trimestre**
  - la **temperatura máxima registrada**
- Imprime cuántas semanas estuvieron **por encima de la media**.

### 📤 Salida de referencia (formato orientativo)

```scala
Top 3 semanas más cálidas:
Sem 8 -> 20.0 ºC
Sem 7 -> 19.1 ºC
Sem 9 -> 18.7 ºC

Temperatura media: 16.63 ºC
Temperatura máxima: 20.0 ºC
Semanas por encima de la media: 6
```


````scala

val temperaturas = List(
  12.5, 13.1, 14.8, 15.2,
  16.0, 18.3, 19.1, 20.0,
  18.7, 17.2, 15.9, 14.4
)

val semanas = List(
  "Sem 1", "Sem 2", "Sem 3", "Sem 4",
  "Sem 5", "Sem 6", "Sem 7", "Sem 8",
  "Sem 9", "Sem 10", "Sem 11", "Sem 12"
)

// Escribe aquí tu solución
val datosSemanasTemperaturas = semanas.zip(temperaturas)
val temperaturaPromedio = temperaturas.sum / temperaturas.length
val temperaturaMaxima = temperaturas.max  

println("TOP Semanas máximas temperaturas:")
datosSemanasTemperaturas.sortBy(-_._2).take(3).foreach{ case (semana, temp) =>
  println(s"$semana: $temp ºC")
}
println(f"Temperatura promedio: $temperaturaPromedio%.2f ºC")
println(f"Temperatura máxima: $temperaturaMaxima%.2f ºC")

println("Semanas por encima del promedio: " + datosSemanasTemperaturas.count(_._2 > temperaturaPromedio))
datosSemanasTemperaturas.filter(_._2 > temperaturaPromedio).foreach{ case (semana, temp) =>
  println(s"$semana: $temp ºC")
}
````

# 🧪 Pregunta 9: Análisis de reservas de hotel con `Map`, `Set` y `Tuple`

Vas a analizar un pequeño dataset de **reservas hoteleras**, representado como una **`List` de `Tuple3`**.

```scala
// Dataset: (habitación, tipo, precio por noche)
val reservas: List[(String, String, Double)] = List(
  ("101", "Individual", 85.0),
  ("102", "Doble", 120.0),
  ("103", "Suite", 250.0),
  ("104", "Doble", 130.0),
  ("105", "Individual", 90.0),
  ("106", "Suite", 275.0),
  ("107", "Doble", 125.0),
  ("108", "Individual", 80.0)
)
```

#### Preguntas

```scala
// 1. Número total de reservas
???

// 2. Tipos de habitación únicos (usando Set)
???

// 3. Precio medio por tipo de habitación (usando Map y groupBy)
???

// 4. Habitación más cara de cada tipo
???

// 5. Estadísticas globales de precios (min, max, media y total) usando Tuple
???
```

### 📤 Salida de referencia (formato orientativo)

```scala
Total de reservas: 8
Tipos de habitación: Set(Individual, Doble, Suite)

=== Precio medio por tipo ===
Individual -> 85.00 €
Doble      -> 125.00 €
Suite      -> 262.50 €

=== Habitación más cara por tipo ===
Individual -> 105
Doble      -> 104
Suite      -> 106

=== Estadísticas globales ===
Precio mínimo: 80.00 €
Precio máximo: 275.00 €
Precio medio: 144.38 €
Facturación total: 1155.00 €
```


````scala

val reservas: List[(String, String, Double)] = List(
  ("101", "Individual", 85.0),
  ("102", "Doble", 120.0),
  ("103", "Suite", 250.0),
  ("104", "Doble", 130.0),
  ("105", "Individual", 90.0),
  ("106", "Suite", 275.0),
  ("107", "Doble", 125.0),
  ("108", "Individual", 80.0)
)

// 1. Número total de reservas
val totalReservas = reservas.length
println(s"Total de reservas: $totalReservas")

// 2. Tipos de habitación únicos (usando Set)
val tiposHabitacionUnicos = reservas.map(_._2).toSet
println(s"Tipos de habitación: $tiposHabitacionUnicos")

// 3. Precio medio por tipo de habitación (usando Map y groupBy)
val preciosPorTipo = reservas
        .groupBy(_._2)
        .view.mapValues(precios => precios.map(_._3).sum / precios.length)
        .toMap
println("\n=== Precio medio por tipo ===")
preciosPorTipo.foreach { case (tipo, precio) =>
  println(f"$tipo%-15s-> ${precio}%.2f €")
}

// 4. Habitación más cara de cada tipo
val habitacionMasCaraPorTipo = reservas
        .groupBy(_._2)
        .view.mapValues(precios => precios.maxBy(_._3))
        .toMap
println("\n=== Habitación más cara por tipo ===")
habitacionMasCaraPorTipo.foreach { case (tipo, (_, _, precio)) =>
  println(f"$tipo%-15s-> ${precio}%.2f €")
}

// 5. Estadísticas globales de precios (min, max, media y total) usando Tuple
val estadisticas = (
  reservas.map(_._3).min,
  reservas.map(_._3).max,
  reservas.map(_._3).sum / reservas.length,
  reservas.map(_._3).sum
)
val preciosMinimo = estadisticas._1
val preciosMaximo = estadisticas._2
val preciosMedia = estadisticas._3
val preciosTotal = estadisticas._4

println("\n=== Estadísticas globales ===")
println(s"Precio mínimo:    ${preciosMinimo}%.2f €")
println(s"Precio máximo:    ${preciosMaximo}%.2f €")
println(s"Precio medio:     ${preciosMedia}%.2f €")
println(s"Facturación total: ${preciosTotal}%.2f €")
````

# 🧪 Pregunta 10: Control de asistencia a actividades

Un centro de formación registra los **alumnos que han asistido a dos actividades distintas** durante la semana. Dispones de **dos `Set[String]`** con los nombres de los alumnos que asistieron a cada actividad.

Calcula e imprime:

- Los alumnos que asistieron a **ambas actividades**
- Los alumnos que asistieron **solo a la actividad A**
- Los alumnos que asistieron **solo a la actividad B**
- El **total de alumnos únicos** que participaron al menos en una actividad

```scala
val actividadA = Set("Ana", "Luis", "Marta", "Carlos", "Elena")
val actividadB = Set("Marta", "Carlos", "Pablo", "Sofía", "Luis")
```

### 📤 Salida esperada
```
Recibieron ambas actividades: Set(Luis, Marta, Carlos)
Solo actividad A:             Set(Ana, Elena)
Solo actividad B:             Set(Pablo, Sofía)
Total alumnos únicos:         7
```


````scala

val actividadA = Set("Ana", "Luis", "Marta", "Carlos", "Elena")
val actividadB = Set("Marta", "Carlos", "Pablo", "Sofía", "Luis")

// Escribe aquí tu solución
val ambasActividades = actividadA.intersect(actividadB)
val soloActividadA = actividadA.diff(actividadB)
val soloActividadB = actividadB.diff(actividadA)
val totalAlumnosUnicos = actividadA.union(actividadB)

println(s"Alumnos en ambas actividades: $ambasActividades")
println(s"Alumnos solo en actividad A: $soloActividadA")
println(s"Alumnos solo en actividad B: $soloActividadB")
println(s"Total de alumnos únicos: ${totalAlumnosUnicos.size}")
````
